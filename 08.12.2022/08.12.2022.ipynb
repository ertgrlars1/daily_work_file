```python
#1
import pandas as pd
import numpy as np
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],  'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left, right, on = 'key', how = 'inner')
```

```python
pd.merge(left, right, on = 'key', how = 'right')
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']}, index = ['K0', 'K1', 'K2'])
left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'], 'D': ['D0', 'D2', 'D3']},index = ['K0', 'K2', 'K3'])
right
```

```python
left.join(right, how = 'inner')
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],  'B': ['B0', 'B1', 'B2', 'B3'], 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],  'B': ['B4', 'B5', 'B6', 'B7'], 'C': ['C4', 'C5', 'C6', 'C7'], 'D': ['D4', 'D5', 'D6', 'D7']})
df2
```

```python
#2
df1 = pd.DataFrame({'ID' : [1, 2, 3, 5, 9],
                    'Col_1' : [1, 2, 3, 4, 5],
                    'Col_2' : [6, 7, 8, 9, 10],
                    'Col_4' : 'apple orange banana strawberry raspberry'.split()})

df2 = pd.DataFrame({'ID' : [1, 1, 3, 5],
                    'Col_A' : [8, 9, 10, 11],
                    'Col_B' : [12, 13, 15, 17],
                    'Col_4' : 'apple orange banana kiwi'.split()})
```

```python
df1
```

```python
df2
```

# merge

```python
pd.merge(df1, df2)
# aynı olan sutundaki aynı elemanları merge yaptı. id ve Col_4 sutunları eşleşti
```

```python
pd.merge(df1, df2, on = 'ID')
# ortak ID'leri aldı
```

```python
pd.merge(df1, df2, on = ['ID', 'Col_4'])
# ilk yaptıgımız merge işlemi ile aynı
# cünkü ortak olan 2 tane sutun var ve zaten onları almıştık
```

```python
pd.merge(df1, df2, suffixes = ['_l', '_r'], left_on = 'Col_2', right_on = 'Col_A')
```

```python
pd.merge(df1, df2, suffixes = ['_l', '_r'], left_index=True, right_index=True)
```

Join

ınner ortak

outer hepsi

left soldakinin hepsi

right sagdakinin hepsi


```python
pd.merge(df1, df2, on = 'Col_4', how = 'outer', suffixes = ['_l', '_r'])
```

```python
pd.merge(df1, df2, on = 'Col_4', how = 'left', suffixes = ['_l', '_r'])
```

```python
pd.merge(df1, df2, on = 'Col_4', how = 'right', suffixes = ['_l', '_r'])
```
