```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (6,4)
plt.rcParams['figure.dpi'] = 70
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

**Bar plots**

# 1. Set darkgrid style from seaborn

```python
# CODE HERE
sns.set_style("darkgrid")
```

# 2. Make a bar plot

```python
# CODE HERE
df.plot(kind = "bar");
```

# 3. Stack the bars

```python
# CODE HERE
df.plot.bar(stacked = True);
```

# 4. Rotate the labels and set figsize

```python
# CODE HERE
df.plot.bar(stacked = True, figsize = (9,6), rot = 0);
```

# 5. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
# CODE HERE
df.index = "Game-1, Game-2, Game-3, Game-4, Game-5, Game-6, Game-7".split()
```

# 6. Unstuck the bars, annotate the hight of the bars on top of them

```python
# CODE HERE
ax = df.plot(kind = "bar", figsize = (9,6))

for i in ax.containers :
    plt.bar_label(i);
```

**lineplots**

```python
tips.head()
```

# 7. Plot the avg tip by size  (lineplot)

```python
# CODE HERE
tips.groupby("size").tip.mean().plot();
```

# 8.Set the linestyle as "--"

```python
# CODE HERE
tips.groupby("size").tip.mean().plot(ls = "--");
```

# 9. Make a scatter plot between tip and total_bill

```python
# CODE HERE
tips.plot(kind = "scatter", x = "tip", y = "total_bill");
```

# 10. Set an additional dimension using size column

```python
# CODE HERE
tips.plot(kind = "scatter", x = "tip", y = "total_bill", colormap = "coolwarm", c = "size")
```

# 11. Make a box plot of total_bill column

```python
# CODE HERE
sns.set_style("white")
tips.total_bill.plot(kind = "box");
```

# 12. Seperate the the boxplot above using size columns

```python
# CODE HERE
tips.boxplot(column="total_bill", by="size");
```

# 13. Make the same plot using seaborn

```python
# CODE HERE
sns.set_style("darkgrid")
sns.boxplot(x = "size", y = "total_bill", data = tips);
```

# 14. Make a violinplot instead of boxplot and discuss the difference between boxplot and violinplot
