```python
#3
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 1
Creat a plot of $\ e^{-x}sin(4x)$ and $\ e^{-x}cos(4x)$ from 0 to 5. 

Change the line color for $\ e^{-x}sin(4x)$ to red, and $\ e^{-x}cos(4x)$ to blue. Adjust the line width to 2.0. Add legend, xlabel, ylabel, and plot title. 

```python
x = np.linspace(0,5,100)
y = np.sin(4 * x) * np.exp(-x)
z = np.exp(-x) * np.cos(4 * x)
plt.figure()
plt.plot(x, y, "r--", label = "$e^{-x} sin(4x)$", linewidth = 4.0)
plt.plot(x, z, label = "$e^{-x} cos(4x)$", linewidth = 2.0)
plt.xlabel('x')
plt.ylabel("value")
plt.legend()
plt.title("My Plot")
plt.savefig("myFigure.png")
```

## Problem 2

Create 4 subplots of $y_1=x$, $y_2=x^2$, $y_3=x^3$, and $y_4=x^{0.5}$. Use 10 points between $x=0$ to $x=10$. For $y_1 = x$ use red circles, $y_2 = x^2$ use green dashes, $y_3 = x^3$ use blue triangles, and $y_4 = x^{0.5}$ use black squares. Also give each subplot a title. Include a command to save the figure as a PNG file so that it can be imported into another program such as Microsoft Powerpoint.

```python
x = np.linspace(0,10,100)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = np.sqrt(x)
plt.figure(figsize=(6, 6), dpi = 100)
plt.subplot(2,2,1)
plt.plot(x, y1, 'r--');
plt.title("$y_1 = x$")

plt.subplot(2,2,2)
plt.plot(x, y2, 'g--');
plt.title("$y_2 = x^2$");


plt.subplot(2,2,3)
plt.plot(x, y3, 'b--');
plt.title("$y_3 = x^3$");

plt.subplot(2,2,4)
plt.plot(x, y4, 'm--');
plt.title("$y_4 = \sqrt{x}$");
plt.tight_layout()
# subplot'lar birbirine yapışmasın diye kullanıldı
plt.savefig("my.png")
```
