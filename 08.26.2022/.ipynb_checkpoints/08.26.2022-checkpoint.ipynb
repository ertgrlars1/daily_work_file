```python
#2
cars = sns.load_dataset('mpg').dropna()
cars = cars[cars.cylinders.isin([4, 6, 8])] 
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year]
cars.head()
```

# Basics

```python
sns.catplot(x='horsepower', data=cars);
# kind default strip
```

```python
sns.catplot(y='horsepower', data=cars);
```

## kind

```python
sns.catplot(x='horsepower', data=cars, kind='box');
```

```python
sns.catplot(x='horsepower', data=cars, kind='violin');
```

## Categorical Variables, Hue

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            kind='box');
```

```python
sns.catplot(x='origin', y='horsepower',
            data=cars, 
            hue='cylinders',
            kind='strip'); 
```

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            hue='cylinders',
            kind='box');
```

# Small Multiples (FacetGrid)

## row, col

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin'
           );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin'
           );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type'
           );
```

## row_order (order, hue_order, col_order)

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type',
            row_order=['new', 'old']
           );
```

```python
#4
geyser = sns.load_dataset("geyser")
geyser.head()
```

# Basics

```python
sns.jointplot(x=geyser.waiting, y=geyser.duration);
# sns.jointplot(x='waiting', y='duration', data=geyser);
```

## Plot Kinds
There are currently six kinds of jointplots allowed in seaborn:

1) scatter (default)

2) kde

3) reg

4) hist

5) hex

6) resid

### scatter (Default)

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='scatter'
             );
```

### kde

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='kde'
             );
```

### reg

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg'
             );
```

### hist

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='hist'
             );
```

# Categorical Variables

```python
geyser.head()
```

### hue

```python
sns.jointplot(x='waiting', y='duration', data=geyser);
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              hue='kind'
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              hue='kind',
              kind='kde'
             );
```

# Styling

## color, palette

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              color='green'
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              hue='kind',
              palette='autumn'
             );
```

## height, space, ratio

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              height=20,   #Default: 6
              space=0.2,  #Default: 0.2
              ratio=20     #Default: 5
             );
#space plot ile histogramlar aras bosluk
# height byklk, inc olarak
# ratio plot'un genlii
```

### joint_kws
Joint keywords should match the figure type in the middle of the jointplot.


```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              joint_kws={'marker': 5}
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg');
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg',
              joint_kws={'ci': None}
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg',
              joint_kws={'ci': None, 'order':3}
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg',
              joint_kws={'ci': None, 'order': 10}
             );
```

### marginal_kws
Marginal keywords should match the type of plot for the marginal distribution plots (the outside plots).

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              marginal_kws={'color': 'xkcd:golden'}
             );
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='kde',
              marginal_kws={'lw': 4, 'shade': True}
             );
```
