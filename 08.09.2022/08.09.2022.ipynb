```python
#3
df = pd.DataFrame({'groups' : 'A B C A B C'.split(),
                  'var1' : [10, 23, 33, 22, 11, 99],
                  'var2' : [100, 253, 333, 262, 111, 969]})
df
```

```python
df.groupby('groups').aggregate([np.min,np.median,np.max])
```

```python
df.groupby('groups').agg(['min', 'median', 'max'])
```

```python
df.groupby('groups').transform(lambda x : x.mean())
```

```python
df.groupby('groups').filter(lambda x : x['var1'].mean() < 17)
```

```python
df.apply(np.sum)
```

```python
#4
df = pd.DataFrame({'key' : 'A B C A B C'.split(),
                  'data1' : range(6),
                  'data2' : [5, 0, 3, 3, 7, 9]})
df
```

```python
df.groupby('key').data1.median()
```

### Aggregation

```python
df.groupby('key').aggregate(['min', np.median, max])
# string gibi,numpy gibi, function gibi yazlablrler
```

```python
df.groupby('key').aggregate({'data1' : 'min',
                             'data2' : 'max'})
```

### Filtering

```python
df.groupby('key').std()
```

```python
df.groupby('key').filter(lambda x : x['data2'].std() > 4)
```

### Transform

```python
# veri setinde 6 satr varsa ckt da 6 satr olacaktr
# her birine matematk ilem uygular
```

```python
df
```

```python
df.groupby('key').transform(lambda x : x - x.mean())
```

### apply()

```python
df
```

```python
def norm_by_data2(x) :
    x['data1'] /= x['data2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_data2)
```
