# Violinplot Options

## legend

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )

plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )

plt.legend(loc=3);
```

## split

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True
              )

plt.legend(loc=2);
# hem japan hem de europe'yi split ile aynı violin'e koydok
```

## inner

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner='quartiles'
              )

plt.legend(loc=2);
# inner quartiles ile .25, .50, .75 ceyreklerının nerede oldugunu gosterıyor
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner=None
              )

plt.legend(loc=2);
```

## scale

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count'
              )

plt.legend(loc=2);
# genişlik scale = "count" ile sahip olunan gözlem sayısına göre ölçeklendi
```

## scale_hue

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count',
               scale_hue=False
              )

plt.legend(loc=2);
#scale_hue=False ile her iki violine göre bir kıyaslama yaptı.
# yani 6 sılındırlı grubun 4 sılındırlı gruptan daha kucuk oldugunu gormus olduk
```
